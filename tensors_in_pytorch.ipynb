**Inplace operation**

```python
n = torch.rand(2,3)
o = torch.rand(2,3)
print(n)
print(o)
```

```python
n+o
#here this creates a new tensor and occupies more space
# so we need another approach to add the tensor and does not use extra space
```

```python
#good pproach
n.add_(o)
```

```python
n
```

```python
o
```

```python
torch.relu(m)
#this is completely new tensor and uses extra space
```

```python
m.relu_()
#this does not uses extra space
```
